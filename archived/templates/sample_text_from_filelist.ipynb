#### How to create a text sample from a list of filenames. This can technically be run against the entire list of 440,000 filenames but will likely crash your notebook. Instead, create a subset for testing. One filename per line.

```python
from urllib import request
```

#### open the filenames file. loop through each line, creating a list of filenames named 'lines'

```python
filenames_path = "filenames-subset.txt"
s3_root = "https://flint-text.s3.amazonaws.com/"
text_sample = ''

with open(filenames_path) as fp:
    lines = [line.rstrip('\n') for line in fp]
```

#### loop over the list, prepend the s3_root to the name to create a URL, read the contents, append to the string 'text_sample'

```python
for line in lines:       
    url = s3_root + line
    response = request.urlopen(url)
    raw_text = response.read().decode('utf8')
    text_sample += raw_text
```

```python
print(text_sample)
```

#### you now have a string named 'text_sample' that you can test regular expressions against or tokenize with nltk. If you need to perform some operation against each file as it's being read in, do that inside the for loop above

```python
#sample usage
import re
match = re.search('McShane, Hilda', text_sample)
print(match)
```
