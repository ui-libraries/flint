```python
from urllib import request
```

```python
filenames_path = "file-names.txt"
s3_root = "https://flint-text.s3.amazonaws.com/"
text_sample = ''

with open(filenames_path) as fp:
    lines = [line.rstrip('\n') for line in fp]
```

```python
for line in lines:       
    url = s3_root + line
    response = request.urlopen(url)
    raw_text = response.read().decode('utf8')
    text_sample += raw_text
```

```python
print(text_sample)
```

```python
import urllib
import os
import re
```

```python
def preprocess(x):
    x = x.replace('\t',' ')
    x = x.replace('\n',' ')
    x = x.replace('(',' ')
    x = x.replace(')',' ')
    x = x.replace('[',' ')
    x = x.replace(']',' ')
    x = x.replace('{',' ')
    x = x.replace('}',' ')
    x = x.replace(',',' ')
    x = x.replace('"','')
    x = x.replace("'",'')
    return(x)
```

```python
reg = '|'.join(DateTimeRegex.values())
```

```python
def DateTimeExtractor(x):
    x = preprocess(x)
    DT = re.findall(reg,x)
    return(DT)
```
