```python
import re
f = open("deq14sample2.txt", "r+")
deqsstr = f.read()
deqslist = re.split("\n", deqsstr)
```

```python
from urllib import request

emailstext = ''

s3_root = "https://flint-text.s3.amazonaws.com/"
for deq in deqslist:       
    url = s3_root + deq
    response = request.urlopen(url)
    raw_text = response.read().decode('utf8')
    emailstext = "".join((raw_text, emailstext)) #changed this line
```

```python
#print(emailstext) #see full email output
```

```python
froms = re.findall("From:.[A-Za-z].*", emailstext)
fromstr = '\n'.join(froms)
fromstr_orgs = re.sub("From:\W","", fromstr) #with (ORG)
fromstr_noorgs = re.sub("[(].*[)]","", fromstr_orgs)#without (ORG)
fromstr_noemail1 = re.sub("[<].*[>]*","", fromstr_noorgs)
fromstr_noemail2 = re.sub("\[.*\]*","", fromstr_noemail1)
print(fromstr_noemail2)
```

```python
roughlist = re.split('\n', fromstr_noemail2) #splits from str into list
onsenders = re.split('\s\n',onstr2) #splits on... str into list
roughlist.extend(onsenders) #combines the two lists
roughstr = '\n'.join(roughlist) #makes list into string
print(roughstr)
len(roughlist)
```

```python
allthenames = re.split('\n',roughstr)
print("There were " + str(len(allthenames)) + " total names found.")
```

```python
ultimatelist = [] #creates final list
```

```python
# extracts sender lines formatted in [Last, First] 
lastcommafirstdude = re.findall("[A-Z][a-z]+.[A-Za-z]+,\s[A-Z][a-z][A-Za-z]+",roughstr)
ultimatelist.extend(lastcommafirstdude) #puts the names already correctly formatted into ultimate list
print(lastcommafirstdude)
len(lastcommafirstdude) #number of names already in LAST, FIRST format
```

```python
# extracts sender lines formatted in [First Last] 
firstlast = re.findall("[A-Z][a-z]+.[A-Z][a-z][A-Za-z]+",roughstr)
#print(firstlast)
```

```python
#problem: for "Shekter Smith, Liane", "Shekter Smith" is read as a [First Last]
#band-aid solution:
unwanted = {'Shekter Smith', 'Mary Beth'}
firstlast1 = [e for e in firstlast if e not in unwanted]
firstlaststring = '\n'.join(firstlast1)
print(firstlaststring)
```

```python
msg = len(allthenames) - (len(lastcommafirstdude) + len(firstlast1)) # number of names not accounted for
if msg > 0:
    print ("rats! you're missing " + str(msg) + ' names.')
if msg == 0:
    print ("Success! all names are accounted for.")
if msg < 0:
    print ("Duplicates present")
```
